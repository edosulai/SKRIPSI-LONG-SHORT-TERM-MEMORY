```python
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
dateparse = lambda x: datetime.strptime(x, '%d-%m-%Y')
df = pd.read_csv('../Data/1985-2021.csv', date_parser=dateparse, parse_dates=['Tanggal'], index_col=['Tanggal'], usecols=['Tanggal', 'Tavg', 'RH_avg', 'RR', 'ff_avg']).replace(to_replace=[8888, 9999, 2555], value=np.nan)
df
```

```python
df['2020-06-12':'2020-06-17']
```

```python
df.dropna()['2020-06-12':'2020-06-17']
```

```python
interpolate = df.interpolate()
interpolate['2020-06-12':'2020-06-17']
```

```python
# columns = ['Tn', 'Tx', 'Tavg', 'RH_avg', 'RR', 'ss', 'ff_x', 'ddd_x', 'ff_avg']
columns = ['Tavg', 'RH_avg', 'RR', 'ff_avg']
scaller = MinMaxScaler(feature_range=(0, 1))
interpolate[columns] = scaller.fit_transform(interpolate[columns])
interpolate['2020-06-12':'2020-06-17']
```

```python
train, test = train_test_split(interpolate, test_size=0.1, shuffle=False)
```
